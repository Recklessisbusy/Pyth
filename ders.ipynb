```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,50,100)
y = np.sin(x)
z = np.cos(x)

plt.subplot2grid((2,4),(0,0), colspan=1)
plt.plot(x,y)

plt.subplot2grid((2,4),(0,1), colspan=1)
plt.plot(x,z)

plt.subplot2grid((2,4),(0,2), colspan=1)
plt.plot(y,z)

plt.subplot2grid((2,4),(1,0), colspan=3)
plt.plot(x,y)

plt.subplot2grid((2,4),(0,3), rowspan=3)
plt.plot(x,y)

plt.show()


```
